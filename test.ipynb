```python
import string

alllines = open('./data/shakespear_alllines.txt').readlines() 
# return a list, each element is a raw string, like '"ACT I"\n'

def proc_line(eachline):
    no_endl = eachline.strip()
    
    rmpunc = str.maketrans('','', string.punctuation) 
    # when passing 3 variables to maketrans, the first 2 are replacement maps, each char in the 3rd string will be deleted
    # string.punctuation is '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~'
    # the maketrans obj has to be used by string.translate fuction, kind of like regular expression
    
    nopunc_all_lower = no_endl.translate(rmpunc).lower()
    return nopunc_all_lower.split()

allwords = []
for eachline in alllines:
    allwords.extend(proc_line(eachline))

def count_words(wordlist):
    worddict = {}
    for eachword in wordlist:
        if eachword in worddict:
            worddict[eachword] += 1
        else:
            worddict[eachword] = 1
    return worddict

worddict = count_words(allwords)

wordcount = len(worddict)
print(f'Shakespeare doc has a total of {wordcount} unique words.')

sortedcount = sorted(worddict.items(), key = lambda x: x[1], reverse = True)

print(f'The most frequent 10 words are: ')
print(*(sortedcount[0:10]), sep ="\n")
```
